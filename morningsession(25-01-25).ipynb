```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
df.dropna(subset=['AverageTemperature'], inplace=True)
avg_temp_by_country = df.groupby('Country')['AverageTemperature'].mean().reset_index()
sorted_avg_temp = avg_temp_by_country.sort_values(by='AverageTemperature', ascending=False)

# bar plot 
top_10_countries = sorted_avg_temp.head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country'], top_10_countries['AverageTemperature'])
plt.xlabel("Country")
plt.ylabel("Average Temperature")
plt.title("Top 10 Countries with Highest Average Temperatures")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print("Top 10 Countries with Highest Average Temperatures:")
top_10_countries
```

```python
#Box plot
#what is spread of temperature for a specific country over a given period
#use case:S ummarize data using quantilers,median,and outliers
#explot:
#x-axis:Years or countries
#Y-axis: Average temperature
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
df.dropna(subset=['AverageTemperature'], inplace=True)
df['dt'] = pd.to_datetime(df['dt'])
df['Year'] = df['dt'].dt.year
country_year_temp = df.groupby(['Country', 'Year'])['AverageTemperature'].mean().reset_index()

#box plot
plt.figure(figsize=(12, 6))
plt.boxplot([country_year_temp[country_year_temp['Country'] == 'India']['AverageTemperature'].values
             for country in country_year_temp['Country'].unique()],
            labels=country_year_temp['Country'].unique())
plt.xlabel("Country")
plt.ylabel("Average Temperature")
plt.title("Temperature Spread by Country")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

c = df[df['Country'] == 'India'].copy()
c['dt'] = pd.to_datetime(c['dt'], errors='coerce')
c = c.dropna(subset=['dt'])
c['Year'] = c['dt'].dt.year
c = c[['Year', 'AverageTemperature']].dropna()

# Group the temperatures by year
g = [group['AverageTemperature'] for _, group in c.groupby('Year')]

# Create the boxplot
plt.figure(figsize=(12,6))
plt.boxplot(g, showfliers=True)  
plt.title("Temperature spread in India")
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.show()
```
