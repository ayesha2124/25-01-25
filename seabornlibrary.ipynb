```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")

sns.barplot(data=df,x="Pclass",y="Survived",palette="rocket_r",errorbar=None)
plt.title("survival rates of passenger")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")

sns.scatterplot(data=df,x="Age",y="Fare",hue='Sex',palette="crest")
plt.title("Relationship between age and fare")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")
s=df[['Age', 'Fare', 'Pclass','Survived']]
sns.pairplot(s,hue='Survived',diag_kind='kde',palette="crest")
plt.title("relation between age fare and p class",x=-1,y=3.3)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")

df_numeric = df.select_dtypes(include=['number'])
s = df_numeric.corr()
sns.heatmap(s, annot=True, cmap="PiYG", fmt=',.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
